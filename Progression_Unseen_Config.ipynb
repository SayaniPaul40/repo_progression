**'Progression_Unseen_Config' notebook:**

This notebook creates a configuration file for the progression model. The config file stores some necessary table names, output tables for unseen prediction and keeps a track of scheduled run of the notebooks.

**Importing required libraries and initializing Spark session**

```python
import json
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("Create JSON").getOrCreate()
```

**This configuration file will be used to track biweekly progression prediction run. Also it contains abcs external data, characteristics table, output file locations generated after the unseen predictions.** 

**JSON file is saved in this Silver layer: _"/lakehouse/default/Files/progression_config_template/progression_config.json"_**

```python
#JSON data structure as a Python dictionary
json_data = {
    "abcs_external":"SilverData.external_progression_data",
    "characteristics_table":"SilverData.event_student_characteristics",
    "unseen_output":{
        "bl_unseen_output_path":"GoldData.progression_unseen_data_prediction_bl",
        "dl_unseen_output_path":"GoldData.progression_unseen_data_prediction_dl"
    },
    "pipeline_schedule":{
        "bl_last_biweekly_run":"2024-11-03",
        "dl_last_biweekly_run":"2024-11-03"
    },
    "progression_log": "SilverData.progression_log_table"   
} 

# Convert the dictionary to a JSON string
json_str = json.dumps(json_data)

# Write the JSON data to a file
file_path = "/lakehouse/default/Files/progression_config_template/progression_config.json"

# Save the JSON string to OneLake or a path available in Microsoft Fabric
with open(file_path, 'w') as json_file:
    json_file.write(json_str)

print(f"JSON file saved at: {file_path}")
```
